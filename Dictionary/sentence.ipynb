```python
from nltk.corpus import wordnet
import pandas as pd
```

```python
similar = pd.read_csv('Similar.csv')
```

```python
similar
```

```python
wordlist = similar['Word']
wordlist
```

```python
def each_word(word):
    synonyms = []
    antonyms = []
    for syn in wordnet.synsets(word):
        for lm in syn.lemmas():
            synonyms.append(lm.name())
    print (set(synonyms))

    for syn in wordnet.synsets(word):
        for lm in syn.lemmas():
            if lm.antonyms():
                antonyms.append(lm.antonyms()[0].name())
    print(set(antonyms))
    return synonyms, antonyms
```

```python
contradictory = {}
similar = {}
for word in wordlist:
    synonyms, antonyms = each_word(word=word)
    similar[word] = synonyms
    contradictory[word] = antonyms
display(similar)
display(contradictory)
```

```python
import csv
with open('test.csv','w',newline='') as f:
    writer = csv.writer(f)
    for i, row in enumerate(similar.items()):
        key = [row[0]]
        values = list(set(row[1]))
        print(key)
        print(values)
        total = key
        if len(values) != 0:
            total.extend(values)
        print(total)
        writer.writerow(total)
```
