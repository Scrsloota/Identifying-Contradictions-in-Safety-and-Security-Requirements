```python
class TimeData:
    def __init__(self, type, time_sentence, period = None, timeline = None, total = None):
        self.type = type # 是否可以在时间线上表示出来
        self.sentence = time_sentence
        self.period = period
        self.timeline = timeline # 起止时间list （start，end)
        self.total = total
```

```python
class State:
    def __init__(self, state_sentence, subject, adjPre=None, time=None):
        self.type = 1
        self.req_id = None
        self.state_sentence = state_sentence
        self.subject = subject
        self.adjPre, self.time = adjPre, time
        self.sentence_index, self.s_index = None, None
    
    def get_req_id(self, req_id):
        self.req_id = req_id
    
    def set_index(self,sentence_index, s_index):
        self.sentence_index,self.s_index = sentence_index,s_index
    
    def show(self):
        print('**state:')
        print(self.type, self.req_id)
        print(self.state_sentence)
        print(self.subject)
        print(self.adjPre)
        if self.time == None:
            print('None')
        else:
            print(self.time.type, self.time.sentence)
```

```python
class Action:
    def __init__(self, action_sentence, subject, verb, object=None, preVerbPhrase=None, adv=None, time=None):
        self.type = 0
        self.req_id = None
        self.action_sentence = action_sentence
        self.subject, self.object = subject, object
        self.verb, self.preVerbPhrase, self.adv, self.time = verb, preVerbPhrase, adv, time
        self.sentence_index, self.s_index = None, None
    
    def get_req_id(self, req_id):
        self.req_id = req_id

    def set_index(self,sentence_index, s_index):
        self.sentence_index,self.s_index = sentence_index,s_index

    def show(self):
        print('**action')
        print(self.type, self.req_id)
        print(self.action_sentence)
        print(self.subject,self.object)
        print(self.verb, self.preVerbPhrase, self.adv)
        if self.time == None:
            print('None')
        else:
            print(self.time.type, self.time.sentence)
```

```python
class Sentence:
    
    def __init__(self, t):
        self.t = t
        self.req_id = None
        self.s = []
        self.sentence_index = None
    
    def get_req_id(self,req_id):
        self.req_id = req_id

    def add_action(self, action:Action):
        action.get_req_id(self.req_id)
        action.set_index(self.sentence_index,len(self.s))
        self.s.append(action)
    
    def add_state(self,state:State):
        state.get_req_id(self.req_id)
        state.set_index(self.sentence_index,len(self.s))
        self.s.append(state)
    
    def set_index(self, sentence_index):
        self.sentence_index = sentence_index
    
    def show(self):
        print('--sentence:')
        print(self.t,self.req_id)
        for p in self.s:
            p.show()
        
```

```python
class Requirement:

    def __init__(self, req_id, requirement_sentence, r):
        self.req_id = req_id
        self.requirement_sentence = requirement_sentence
        self.r = r
        assert self.r in [0, 1]
        self.sentence = []

    def add_sentence(self, excution: Sentence, condition = None):
        excution.get_req_id(self.req_id)
        self.sentence.append(excution)
        if condition is not None:
            condition.get_req_id(self.req_id)
            self.sentence.append(condition)

    def show(self):
        print('----------------------------------------------------------------------------')
        print('--requirement:')
        print(self.r,self.req_id)
        print(self.requirement_sentence)
        for p in self.sentence:
            p.show()

```

```python
def group_elements(s_as):
    Ga = {}
    Gs = {}
    for element in s_as:
        type = element.type
        if type == 0:  # action
            sub = element.subject
            obj = element.object
            if (sub, obj) not in Ga:
                Ga[(sub, obj)] = [element]
            else:
                Ga[(sub, obj)].append(element)
        else:  # state
            sub = element.subject
            if sub not in Gs:
                Gs[sub] = [element]
            else:
                Gs[sub].append(element)
    return list(Ga.values()), list(Gs.values())
```

```python
import numpy as np
similar = {}
contradictory = {}
similar_word = pd.read_csv('Similar.csv')
contradictory_word = pd.read_csv('Contradictory.csv')
similar_list = similar_word.values.tolist()
for i,li in enumerate(similar_list):
    while np.nan in li:
        similar_list[i].remove(np.nan)
    similar[similar_list[i][0]] = list(set(similar_list[i]))
contradictory_list = contradictory_word.values.tolist()
for j, lj in enumerate(contradictory_list):
    while np.nan in lj:
        contradictory_list[j].remove(np.nan)
    contradictory[contradictory_list[j][0]] = list(set(contradictory_list[j]))
display(similar)
display(contradictory)
```

```python
def is_similar(x1, x2):
    if x1 is None or x2 is None:
        return True

    if x1 in similar and x2 in similar[x1] and x2 in similar and x1 in similar[x2]:
        return True
    return False
```

```python
def is_contradictory(x1, x2):
    if x1 is None or x2 is None:
        return False
    if x1 in contradictory and x2 in contradictory[x1] and x2 in contradictory and x1 in contradictory[x2]:
        return True
    return False
```

```python
def relation_of_actions_without_time(action1: Action, action2: Action):
    #判断行为冲突，还有时间冲突，考虑到同时满足才返回对应情况与结果
    not_contra, contra, sim= 0, 1, 2
    if is_similar(action1.verb, action2.verb):
        if is_similar(action1.preVerbPhrase, action2.preVerbPhrase):
            if is_similar(action1.adv, action2.adv):
                return sim
            else:
                return contra
        elif is_contradictory(action1.preVerbPhrase, action2.preVerbPhrase):
            return contra
        else:
            return sim
    elif is_contradictory(action1.verb, action2.verb):
        return contra
    else:
        return not_contra
```

```python

def relation_of_states_without_time(state1: State, state2: State):
    not_contra, contra, sim =0, 1, 2
    if is_similar(state1.adjPre, state2.adjPre):
            return sim
    elif is_contradictory(state1.adjPre, state2.adjPre):
        return contra
    else:
        return not_contra
```

```python
def lcm(x, y):
    a = x
    b = y
    while y:
        x,y = y, x % y
    return a * b // x
```

```python
def relation_of_time(time1, time2):
    #                          存在重合        不重合
    # 时间线明确-时间线明确：  contra          not contra
    # 时间线明确-时间线不明：  潜在冲突         潜在冲突
    # 时间线不明-时间线不明：  潜在冲突         潜在冲突
    not_overlap, overlap, suspec_overlap = 0, 1, 2
    if time1 == None or time2 == None:
        return overlap
    if time1.type == 0 or time2.type == 0:
        return suspec_overlap
    else:
        #（始于当下） period list[start,end] total 判断total和共有周期 然后在周期内按开始时间排序 sort得出重复情况
        period1 = time1.period
        period2 = time2.period
        period = lcm(period1,period2)
        mult1 = period/period1
        mult2 = period/period2
        timeline1 = time1.timeline
        timeline2 = time2.timeline
        total = min(time1.total, time2.total)
        for i in range(1,mult1):
            for pair in time1.timeline:
                start = pair[0]+period1*i
                if start> total:
                    break
                end = pair[1]+period1*i
                if end > total:
                    end = total
                timeline1.append((start,end))
        for i in range(1,mult2):
            for pair in time2.timeline:
                start = pair[0]+period2*i
                if start> total:
                    break
                end = pair[1]+period2*i
                if end > total:
                    end = total
                timeline2.append((start,end))
        timeline = timeline1
        timeline.extend(timeline2)
        timeline.sort()
        total =  time1.total
        for i in range(len(timeline)):
            if i+1 == len(timeline):
                break
            if timeline[i][1]>= timeline[i+1][0]:
                return overlap
        return not_overlap
```

```python
def relation_of_actions(action1:Action, action2:Action):
    not_contra, contra, sim, suspec = 0,1,2,3
    action_result = relation_of_actions_without_time(action1 = action1, action2=action2)
    time_result = relation_of_time(action1.time,action2.time)
    if action_result == 1 and time_result == 1:
        return contra
    elif action_result == 1 and time_result == 2:
        return suspec
    elif action_result == 2:
        return sim
    else:
        return not_contra
```

```python
def relation_of_states(state1:State, state2:State):
    not_contra, contra, sim, suspec = 0,1,2,3
    action_result = relation_of_states_without_time(state1=state1, state2=state2)
    time_result = relation_of_time(state1.time,state2.time)
    if action_result == 1 and time_result == 1:
        return contra
    elif action_result == 1 and time_result == 2:
        return suspec
    elif action_result == 2:
        return sim
    else:
        return not_contra
```

```python
def Judge_actions_and_states(Ga, Gs):
    A_contra = []
    A_sim = []
    A_suspected = []
    Relation_between_req_and_s={}
    for group in Ga:
        for i in range(len(group)):
            for j in range(i + 1, len(group)):
                result = relation_of_actions(group[i], group[j])
                if result == 1:
                    A_contra.append(tuple(sorted([group[i].req_id, group[j].req_id])))
                elif result == 2:
                    req_id_pair = tuple(sorted([group[i].req_id, group[j].req_id]))
                    A_sim.append(req_id_pair)
                    if req_id_pair not in Relation_between_req_and_s:
                        Relation_between_req_and_s[req_id_pair] = [(group[i],group[j])]
                    else:
                        Relation_between_req_and_s[req_id_pair].append((group[i],group[j]))
                elif result == 3:
                    A_suspected.append(tuple(sorted([group[i].req_id, group[j].req_id])))
    S_contra = []
    S_sim = []
    S_suspected = []
    for group in Gs:
        for i in range(len(group)):
            for j in range(i + 1, len(group)):
                result = relation_of_states(group[i], group[j])
                if result == 1:
                    S_contra.append(tuple(sorted([group[i].req_id, group[j].req_id])))
                elif result == 2:
                    req_id_pair = tuple(sorted([group[i].req_id, group[j].req_id]))
                    S_sim.append(req_id_pair)
                    if req_id_pair not in Relation_between_req_and_s:
                        Relation_between_req_and_s[req_id_pair] = [(group[i],group[j])]
                    else:
                        Relation_between_req_and_s[req_id_pair].append((group[i],group[j]))
                elif result == 3:
                    S_suspected.append(tuple(sorted([group[i].req_id, group[j].req_id])))
    R_contra = set(A_contra+S_contra)
    R_sim = set(A_sim+S_sim)-R_contra
    R_suspected = set(A_suspected+S_suspected)-R_contra-R_sim
    Relation_keep = []
    for pair in R_sim:
        Relation_keep.extend(Relation_between_req_and_s[pair])
    return R_contra,R_sim,R_suspected,Relation_keep
```

```python
def Classification(Relation_keep):
    class_list = []
    for x1,x2 in Relation_keep:
        index = -1
        remove_list = []
        for i,s in enumerate(class_list):
            if x1 in s and x2 in s:
                index = i
                break
            elif x1 in s or x2 in s:
                if index == -1:
                    class_list[i].add(x1)
                    class_list[i].add(x2)
                    index = i
                else:
                    class_list[index].add(x1)
                    class_list[index].add(x2)
                    class_list[index] = class_list[index]|class_list[i]
                    remove_list.append(i)
        for ind in reversed(remove_list):
            class_list.pop(ind)
        if index == -1:
            class_list.append(set([x1,x2]))
    return class_list
```

```python
def sim_sequence_list(class_list,sentence_ind):
    sequence = []
    for req in Requirements:
        if req.r == 0:
            sequence.append([])
            continue
        length = len(req.sentence[sentence_ind].s)
        list= [0 for x in range(length)]
        sequence.append(list)
    for i,s in enumerate(class_list):
        for ele in s:
            req_index = ele.req_id
            s_index = ele.s_index
            sequence[req_index][s_index] = i
    return sequence
```

```python
def relation_of_orders(i, j, sequence, sentence_index):
    not_contra, contra, sim, suspec = 0,1,2,3
    task1 = Requirements[i].sentence[sentence_index].t
    task2 = Requirements[j].sentence[sentence_index].t
    if (task1 == 1 and task2 == 2) or (task1 == 2 and task2 == 1):
        return contra
    list1 = sequence[i]
    list2 = sequence[j]
    combine_set = set(list1).intersection(set(list2))
    if 0 in combine_set:
        combine_set.remove(0)
    id1, id2 = 999999,999999
    for letter in combine_set:
        tmp = list1.index(letter)
        if tmp < id1:
            id1 = tmp
        tmp = list2.index(letter)
        if tmp < id2:
            id2 = tmp
    number = len(combine_set)
    if number == 0:
        return not_contra
    p = 0
    zero_flag = False
    while p < number:
        if id1 == len(list1) or id2 == len(list2):
            break
        if list1[id1] == 0:
            zero_flag = True
            id1 = id1 + 1
            continue
        if list2[id2] == 0:
            zero_flag = True
            id2 = id2 + 1
            continue
        if list1[id1] == list2[id2]:
            p = p + 1
            id1 = id1 + 1
            id2 = id2 + 1
        else:
            return contra
    if zero_flag == True:
        return suspec
    return sim
```

```python
def Judge_order(R_sim, sequence,sentence_index):
    O_contra = []
    O_sim = []
    O_suspected = []
    for i,j in R_sim:
        result = relation_of_orders(i, j,sequence,sentence_index)
        if result == 1:
            O_contra.append(tuple(sorted([i,j])))
        elif result == 2:
            O_sim.append(tuple(sorted([i,j])))
        elif result == 3:
            O_suspected.append(tuple(sorted([i,j])))
    return set(O_contra),set(O_sim),set(O_suspected)
            
```

```python
def IdentifyConflicts(Ga,Gs, sentence_index):
    R_contra,R_sim,R_suspected,Relation_keep = Judge_actions_and_states(Ga=Ga,Gs=Gs)
    class_list = Classification(Relation_keep=Relation_keep)
    sequence = sim_sequence_list(class_list=class_list, sentence_ind=sentence_index)
    O_contra, O_sim, O_suspected = Judge_order(R_sim,sequence,sentence_index=sentence_index)
    R_contra = R_contra|O_contra
    R_suspected = R_suspected|O_suspected
    R_sim = O_sim

    return R_contra, R_sim, R_suspected
```

```python
def Judge_condition2(R):
    T_same = []
    Ta = []
    Ts = []
    for i,j in R:
        r1 = Requirements[i].r
        r2 = Requirements[j].r
        if r1 == 0 or r2 == 0:
            T_same.append(tuple(sorted([i,j])))
        else:
            current_set = set()
            req1 = Requirements[i]
            req2 = Requirements[j]
            for s in req1.sentence[1].s:
                current_set.add(s)
            for s in req2.sentence[1].s:
                current_set.add(s)
            Ca, Cs = group_elements(current_set)
            Ta.extend(Ca)
            Ts.extend(Cs)
    T_contra, T_sim, T_suspected = IdentifyConflicts(Ta, Ts, 1)
    return T_contra, T_sim| set(T_same), T_suspected
```

```python
excution_as = set([])
total = []
for first in range(len(Requirements)):
    for second in range(first,len(Requirements)):
        total.append((first, second))
total = set(total)
for req in Requirements:
    for s in req.sentence[0].s:
        excution_as.add(s)
Ga, Gs = group_elements(excution_as)
Contra,Sim,Suspected = IdentifyConflicts(Ga, Gs, 0)
C_T_contra, C_T_sim, C_T_suspected = Judge_condition2(Contra)
S_T_contra, S_T_sim, S_T_suspected = Judge_condition2(Suspected)
Final_contra = C_T_sim & Contra
Final_suspected = ((total-C_T_contra-C_T_sim) & Contra) |(total-S_T_contra & Suspected)
```
